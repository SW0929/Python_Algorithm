## COVID-19 백신 종류별 접종 인구

```python
days = [1, 2, 3] # 1일, 2일, 3일
az = [2, 4, 8] # (단위 : 만명) 1일부터 3일까지 아스트라제네카 접종인구
prizer = [5, 1, 3] # 화이자
moderna = [1, 2, 5] # 모더나

plt.plot(days, az)
plt.plot(days, prizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label='az')
plt.plot(days, prizer, label='prizer', marker='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')

plt.legend()
```

```python
plt.plot(days, az, label='az')
plt.plot(days, prizer, label='prizer', marker='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')

plt.legend(ncol=3)
```
